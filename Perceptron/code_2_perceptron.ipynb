```python
from sklearn.datasets import make_classification
import numpy as np
X,y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes = 2, n_clusters_per_class = 1,random_state=41 , hypercube=False , class_sep=10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=100)
```

```python
#  help to move the perceptone line 
# we have made the perceptron not used the one from sklearn
def perceptron(X,y):
    X=np.insert(X, 0, 1, axis=1)  # Add bias term
    weights = np.ones(X.shape[1])  # Initialize weights
    lr =0.001 # Learning rate

    for i in range(100):  # Number of iterations
        j=np.random.randint(0,100)  # Randomly select an example
        y_hat = step(np.dot(X[j], weights))  # Predict
        weights += lr * (y[j] - y_hat) * X[j]  # Update weights
    return weights[0], weights[1:]
```

```python
def step(z): #activation function
    return 1 if z >= 0 else 0  # Step function for binary classification
    
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print("Intercept:", intercept_)
print("Coefficients:", coef_)
```

```python
m = -(coef_[0] / coef_[1])  # Slope of the line
b = intercept_ / coef_[1]  # y-intercept of the line
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m * x_input + b
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_input, y_input, color='red', linewidth=3)  # Plot
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=100)
plt.ylim(-3, 2)
```
